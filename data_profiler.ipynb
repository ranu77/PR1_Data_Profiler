Note: The Telco Customer Churn dataset does not contain Age, Income, and Purchase columns. Therefore, equivalent variables available in the dataset were used for analysis:

Age - Tenure,
Income - MonthlyCharges,
Purchases - TotalCharges

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["tenure"], kde=True)

plt.title("Distribution of Customer Tenure")
plt.xlabel("Tenure (Months)")
plt.ylabel("Frequency")

plt.show()
```

The tenure distribution shows that a large number of customers have lower tenure values, indicating many customers are relatively new to the service.

```python
plt.figure(figsize=(8,5))

sns.histplot(df["MonthlyCharges"], kde=True)

plt.title("Distribution of Monthly Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Frequency")

plt.show()
```

Monthly Charges are distributed across different pricing plans. The presence of multiple peaks suggests that customers subscribe to different combinations of services, resulting in varying monthly costs.

```python
plt.figure(figsize=(8,5))

sns.histplot(df["TotalCharges"], kde=True)

plt.title("Distribution of Total Charges")
plt.xlabel("Total Charges")
plt.ylabel("Frequency")

plt.show()
```

Total Charges show a right-skewed distribution. Most customers have lower cumulative charges, while a smaller number of long-term customers have accumulated significantly higher charges.

Note: The Telco Customer Churn dataset does not contain Purchases. Therefore, equivalent variable available in the dataset were used for analysis:

Purchases - Churn

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="gender",
    hue="Churn",
    data=df
)

plt.title("Gender vs Churn")
plt.xlabel("Gender")
plt.ylabel("Frequency")

plt.show()
```

The churn distribution appears similar for both male and female customers, suggesting that gender alone may not be a strong predictor of churn.

Note: The Telco Customer Churn dataset does not contain Income. Therefore, equivalent variable available in the dataset were used for analysis:

Income - Monthly Charges

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Churn",
    y="MonthlyCharges",
    data=df
)

plt.title("Monthly Charges vs Churn")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")

plt.show()
```

Customers who churned generally have higher monthly charges compared to customers who stayed. This suggests that higher service costs may contribute to customer churn.

```python
df = pd.read_csv("Customers.csv")
corr_matrix = df.select_dtypes(include='number').corr()
corr_matrix
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.2f',
    cmap='viridis'
)

plt.title('Correlation Heatmap - Telco Customer Churn Dataset')

plt.tight_layout()
plt.show()
```

The correlation heatmap shows relationships among numerical variables. TotalCharges has a strong positive correlation with tenure, indicating that customers who stay longer tend to accumulate higher charges.

```python
sns.pairplot(
    df[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']],
    hue='Churn'
)

plt.suptitle(
    'Pair Plot by Churn',
    y=1.02
)

plt.show()
```

The pair plot shows pairwise relationships between numerical variables and highlights customer churn using different colors. It helps identify patterns, correlations, and differences between churned and non-churned customers.
